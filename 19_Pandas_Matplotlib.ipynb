# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
a = np.random.randint(1,100,60).reshape(20,3)
df = pd.DataFrame(a,columns=["A","B","C"])
df
```

# Line plot

```python
plt.figure(figsize=(16,4))
plt.plot(range(20),df.A)
plt.plot(range(20),df.B)
plt.plot(range(20),df.C)
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.plot(range(20),df)
plt.show()
```

```python
plt.figure(figsize=(16,4))
# plt.plot(range(20),df.A,color="r")
plt.plot(range(20),df.B,"--b")
plt.plot(range(20),df.A,"-r")
plt.show()
```

# scatter plot

```python
plt.figure(figsize=(16,4))
plt.scatter(range(20),df.B,c="r",marker="d")
plt.show()
```

# Bubble chart

```python
plt.figure(figsize=(16,4))
for idx in range(df.B.size):
    #plt.scatter(idx,df["B"][idx],c="r")
#     plt.scatter(idx,df["B"][idx],s=.2*np.pi*df["B"][idx]*df["B"][idx]) # s = 3.14*r2
    plt.scatter(idx,df["B"][idx],s=20*df["B"][idx],alpha=.4) # s = 3.14*r2
plt.show()
```

# Write labels to points

```python
plt.figure(figsize=(16,4))
for idx in range(df.B.size):
    if df["B"][idx] >=80:
        plt.scatter(idx,df["B"][idx],s=20*df["B"][idx],marker="*",color="r")    
        plt.text(idx+.5,df["B"][idx]+2,(idx,df["B"][idx]))
    else:
        plt.scatter(idx,df["B"][idx],s=20*df["B"][idx],alpha=.4)    
plt.show()
```

```python
plt.figure(figsize=(16,4))
for idx in range(df.B.size):
    if df["B"][idx] >=80:
        plt.scatter(idx,df["B"][idx],s=20*df["B"][idx],marker="*",color="r")    
        plt.text(idx+.5,df["B"][idx]+2,(idx,df["B"][idx]))
    elif df["B"][idx] <80 and df["B"][idx] >=40:
        plt.scatter(idx,df["B"][idx],s=5*df["B"][idx],alpha=.4,marker="D")    
    else:
        plt.scatter(idx,df["B"][idx],s=20*df["B"][idx],alpha=.4,marker="^")    

plt.show()
```

# bar chart

```python
plt.figure(figsize=(16,4))
plt.bar(range(20),df.B,color="r")
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.bar(range(20),df.B,color="r")
plt.plot(range(20),df.B,color="r")
plt.stem(range(20),df.B,color="c")
plt.show()
```

# labels,legend and titles

```python
plt.figure(figsize=(16,4))
for idx in range(df.B.size):
    plt.bar(idx,df["B"][idx],color="b")    
    plt.text(idx,df["B"][idx]+1,df["B"][idx])
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.bar(range(20),df["A"],color="c",label="A bar")
plt.bar(range(20),df["B"],color="m",label="B bar")
plt.bar(range(20),df["C"],color="r",label="C bar")
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar graph")
plt.xticks(range(20))
plt.yticks(range(0,100,5))
plt.show()
```
